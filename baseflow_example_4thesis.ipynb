```python
import pandas as pd
```

```python
WOL4 = pd.read_csv('C:/Users/carsk/OneDrive - KU Leuven/Thesis/Revised_RESULTS/Baseflow/WOL4-C01/1H_resampl.csv', index_col=0, parse_dates=True)
WOL4.head()
```

```python
#plot time series of C01_without_baseflow and WOL4

import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.plot(WOL4.index, WOL4['C01_without_baseflow'], label='C01_without_baseflow')
plt.plot(WOL4.index, WOL4['C01'], label='C01')
plt.legend()
plt.show()

```

```python
import plotly.graph_objects as go

fig = go.Figure()

# Add the C01 line
fig.add_trace(go.Scatter(
                x=WOL4.index,
                y=WOL4['C01']/1000,
                name='C01',
                line=dict(color='rgb(187,126,93)', width=2)))

# Add the C01_without_baseflow line
fig.add_trace(go.Scatter(
                x=WOL4.index,
                y=WOL4['C01_without_baseflow'],
                name='C01 without baseflow',
                line=dict(color='rgb(8,146,165)', width=2)))

# Edit the layout
fig.update_layout(
    title='Time series plot of C01 with and without baseflow',
    xaxis_title='Date',
    yaxis_title='Discharge (m3/s)',
    plot_bgcolor='rgb(242,242,242)',  # Light gray background
    legend=dict(
        x=0.75,  # x position of legend (0-1, left to right)
        y=0.99,  # y position of legend (0-1, bottom to top)
        bgcolor='rgba(255,255,255,0.5)',  # semi-transparent white background
        bordercolor='Black',
        borderwidth=1
    )
)

fig.show()
```
